```python
#Importamos os modulos !
import pandas as pd
import pandera as pa
```

```python
#Colocamos aqui os valoresnão informados dentro de umavariavel!
valores_validados = ['**', '****', '###!', '####', '****', '*****', 'NULL']
#e colocamos na leitura do arquivo já para limpar os dados os valores NA_VALUES (Not Avaliable)
dataframe = pd.read_csv('ocorrencia1.csv',sep=';', parse_dates=['ocorrencia_dia'],dayfirst = True, na_values= valores_validados)
```

```python
dataframe.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns={
        "codigo": pa.Column(pa.Int,required = False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
    }
)
```

```python
schema.validate(dataframe)
```

>Ocorrencias com INCIDENTE GRAVE!<

```python
#realizando um filtro nos dados parte2!
filtro1= dataframe.ocorrencia_classificacao == 'INCIDENTE GRAVE'
dataframe.loc[filtro1, ['ocorrencia_classificacao','ocorrencia_cidade', 'total_recomendacoes']]
```

#Combinando 2 filtros!
>Combinando acidentes graves com ESTADO de são paulo!
Combinação feita com &<

```python
filtro2=dataframe.ocorrencia_classificacao =='INCIDENTE GRAVE'
filtro3=dataframe.ocorrencia_cidade =='SÃO PAULO'
dataframe.loc[filtro2&filtro3]
```

#Incidente grave ou ser do estado de SAO PAULO
>Usando o | <

```python
filtro = dataframe.ocorrencia_classificacao =="INCIDENTE GRAVE"
filtro2 = dataframe.ocorrencia_cidade =='SÃO PAULO'
dataframe.loc[filtro | filtro2]
```

#Incidentes graves ou incidentes e o estado SAO PAULO!
>Combinando três fatores para o filtro!
usando no filtro 1 a condição o | (significa ou) e na hora de finalizar usamos a combinação dos dois filtros com o &<

```python
filtro1 = (dataframe.ocorrencia_classificacao =='INCIDENTE GRAVE') | (dataframe.ocorrencia_classificacao =='INCIDENTE')
filtro2 = dataframe.ocorrencia_cidade =="SÃO PAULO"
dataframe.loc[filtro1 &filtro2]
```

>Podemos tambem usar o metodo isin()!
usando como argumento uma lista com os elementos no qual quer colocar na condição!
Veja abaixo um exemplo :<

```python
filtro1 = dataframe.ocorrencia_classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])
filtro2 = dataframe.ocorrencia_uf =='SP'
dataframe.loc[filtro1 & filtro2]
```

- Ocorrencias cuja cidade comecem coma letra C!
>Para isso precissamos obter o primeiro caracter de cadacidade, depois comparar essecaracter com a letra C!<

```python
filtro =dataframe.ocorrencia_cidade.str[0] =='C'
dataframe.loc[filtro]
```

*ocorrencia cuja a cidade terminam com a letra A!
>Para isso usamos o mesmo modelo acima, localizamoso ultimocaracter e comparamos com a letra A<

```python
filtro = dataframe.ocorrencia_cidade.str[-1] =="A"
dataframe.loc[filtro]
```

*Ocorrencias cuja a cidade terminam com MA!
>Para isso podemos usar o mesmo modelo de antes porém usamos uma tecnica de fatiamento da string!e ai comparamos com a silaba MA<

```python
filtro = dataframe.ocorrencia_cidade.str[-2:] =='MA'
dataframe.loc[filtro]
```

*Podemos ainda usar o metodo contains()!
>Usado para verificar se uma condição passada a ele está dentro do objeto a ser verificado!
Nocaso abaixo tentamos passar o mesmo parametro de verificação d a silaba MA dentro da ocorrencia_cidade, porém averificação não é feita apenas no final das palavras!<

```python
filtro= dataframe.ocorrencia_cidade.str.contains('MA')
dataframe.loc[filtro]
```
