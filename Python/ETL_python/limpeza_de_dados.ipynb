```python
#Aqui dizemos que na linha de label 0 na coluna de ocorrencia_aerodromo ela recebe o vazio ''
dataframe.loc[0, 'ocorrencia_aerodromo']= '****'
```

```python
#Pesquisamos dentro do dataframe na coluna ocorrencia_aerodromo objetos que possuam o valor **** 
#e passamos uma função do Pandas para valloresnot avalieble!
dataframe.loc[dataframe.ocorrencia_aerodromo == "****", ['ocorrencia_aerodromo']] = pd.NA
```

```python
dataframe.head()
```

#Caution!
>Fazer a mudança de valores do jeito que está acima, nescessita que seja feito uma a uma para cada valor não informado dentro do dataframe, usaremos a função replace para procurar e modificar eses valores tendo a em ciencia que, para mudar valores não informados do dataframe precisa se consultar quem entende donegocio sendo desenvolvido pela equipe, só se muda algum valor desses játendo o conhecimento de quais são os simbolos/valores substituindoa falta da informação!<

```python
dataframe.replace(['**', '****', '###!', '####', '****', '*****', 'NULL'], pd.NA , inplace=True)
```

```python
dataframe.tail()
```
